### HTML of a basic web page

Let’s say you were asked to find the name and salary of players in a National Basketball League from the following web page. The web page is comprised of HTML. It consists of text surrounded by a series of blue text elements enclosed in angle brackets called <code>tags</code>. The <code>tags</code> tells the browser how to display the content. The data we require is in this text.  

* The first portion contains the <code>"DOCTYPE html”</code> which declares this document is an HTML document. <code> < html > </code>  element is the root element of an HTML page, and <code> < head > </code> element contains meta information about the HTML page. 
* Next, we have the <code> body</code>, this is what's displayed on the web page. This is usually the data we are interested in, we see the <code>elements </code> with an <code> "h3" </code>, this means type 3 heading which makes the text larger and bold. These tags have the names of the players. Notice the data is enclosed in the elements. It starts with a h3 in brackets and ends in a slash h3 in brackets. There is also a different tag <code> “p” </code>, this means paragraph, each <code> p</code> tag contains a player's salary.

### HTML Tree

Each HTML document can actually be referred to as a <code> document tree </code>. Let's go over a simple example:

Tags may contain strings and other tags. These elements are the tag’s children. We can represent this as a <code>family tree</code>. Each nested tag is a level in the tree. The tag HTML tag contains the head and body tag. The Head and body tag are the descendants of the html tag. In particular they are the <code>children</code> of the HTML tag. HTML tag is their parent. The <code>head</code> and <code>body</code> tag are siblings as they are on the same level. <code>Title</code> tag is the child of the <code>head</code> tag and its parent is the <code>head</code> tag. The <code>title</code> tag is a descendant of the <code>HTML</code> tag but not its child. The heading and paragraph tags (<code>< h3 >, < p > </code>) are the children of the <code>body</code> tag; and as they are all children of the <code>body</code> tag they are siblings of each other. The <code>bold</code> tag is a child of the <code>heading</code> tag, the content of the tag is also part of  the tree but this can get unwieldy to draw.

### HTML Tables

<code>< td ></code> tags, each defines a table cell.
