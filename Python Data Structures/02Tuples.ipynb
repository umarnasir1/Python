## Indexing 

Each element of a tuple can be accessed via an index. The following table represents the relationship between the index and the items in the tuple. Each element can be obtained by the name of the tuple followed by a square bracket with the index number:

```python
tuple1 = ('disco', 10, 1.2)

print(tuple1[0]) #disco -Access the element, with respect to index 0
print(tuple1[1]) #10
print(tuple1[2]) #1.2 - Access the element, with respect to index 2

# Print the type of value on each index
print(type(tuple1[0])) #<class 'str'>
print(type(tuple1[1])) #<class 'int'>
print(type(tuple1[2])) #<class 'float'>

# Negative indexing 

#last element
tuple1[-1] #1.2

# Use negative index to get the value of the second last element
tuple1[-2] #10

# Use negative index to get the value of the third last element
tuple1[-3] #'disco'
```

```python
# Uses of Tuple
# Often use to store related pieces of information
# Some times, we need to be sure that the object doesnt change, then we use Tuple and a Tuple guarentees that
# Tuples are usefule when you have two or more values that are so closely related that they will always be used together.
#   e.g. latitude & longitude 
# Tuples can also be used to assign multiple variables in a compact way.

# Eg.1
#Tuples are often used for functions that have multiple return values.
#   For example, the as_integer_ratio() method of float objects returns a numerator and a denominator in the form of 
#   a tuple:
x = 0.125
x.as_integer_ratio() #(1, 8)

# These multiple return values can be individually assigned as follows:
numerator, denominator = x.as_integer_ratio()
print(numerator) # 1
print(denominator) #8

# Eg. 2
location = (13.4125, 103.866667)
print("Latitude:", location[0]) #print("Location is at latitude: {}.".format(location[0]))
print("Longitude:", location[1])

# Eg. 3 -Tuples can also be used to assign multiple variables in a compact way.
length, width, height = 52, 40, 100 #'Tuple unpacking' three variables are assigned from the content of the tuple 
#  dimentions. You can use tuple unpacking to assign the information from a tuple into multiple variables without having
#  to access them one by one and make multiple assignment statements.
print("The dimensions are {} x {} x {}".format(length, width, height)) #The dimensions are 52 x 40 x 100
```
