# Sets
* A set is a data type/container for mutable unordered collections of unique elements.
* As not ordered so there is no last element/no sequence of position as they are unordered & 
    it cannot be indexed and sliced.
* A set cannot contain duplicate elements / only have unique elements -
* Creating a set 
    * converting a list to set - type casting 
    * creating using {}
* Set Operations
* Methods - to modifiy sets 
    * add - add elements in set
    * pop - remove elements from set (a random element is removed) 
    * remove - remove a particular element 
* supports Membership Operators (in, not in)
* Applications:
    * Quickly remove duplicates from a list
    * Sets Logic Operations - operations of methematical sets
        * intersection
        * difference
        * union
        * superset
        * subset

```python
# Defining a set - converting a list to a set 
# Ex 1

numbers = [1, 2, 6, 3, 1, 1, 6]
unique_nums = set(numbers) # Type casting  OR set([1, 2, 6, 3, 1, 1, 6])
type(unique_nums) #set
print(unique_nums) #{1, 2, 3, 6} - no duplicate elements 

# another exemple of set removing duplicate items
A = [1, 2, 2, 1]  
B = set([1, 2, 2, 1])
print("the sum of A is:", sum(A)) #6
print("the sum of B is:", sum(B)) #3
```

```python
# Defining a set 
# Ex 2
set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1 #{'R&B', 'disco', 'hard rock', 'pop', 'rock', 'soul'}
```

## Set Operations
* methods 
    * add 
    * pop 
    * remove 

```python
# add
fruit = {"apple", "banana", "orange", "grapefruit"}
fruit.add("watermelon")  # add an element
fruit #{'banana', 'orange', 'grapefruit', 'apple', 'watermelon'}

#if we add same element twice, nothing will happen 
fruit.add("watermelon")
fruit #{'apple', 'banana', 'grapefruit', 'orange', 'watermelon'}
```

```python
# pop -remove a random element - as there is no last elememnt as sets are unordered 

print(fruit.pop()) #banana 
print(fruit) #{'orange', 'grapefruit', 'apple', 'watermelon'}
```

```python
# remove - removes a particular element

fruit = {"apple", "banana", "orange", "grapefruit"}
fruit.remove("orange")
fruit #{'apple', 'banana', 'grapefruit'}
```

```python
# Verify if the element is in the set

# We can verify if an element is in the set using the in command as follows. 
# If the item is in the set, it returns true.
# If we look for an item that is not in the set, we will get a false.

fruit = {"apple", "banana", "orange", "grapefruit"}  # define a set
print("watermelon" in fruit) # False
```

## Sets Logic Operations

```python
# Remember that with sets you can check the difference between sets, as well as the symmetric difference, intersection,
#   and union:

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

```python
# intersection - finds the elements that are in both sets.

{'a','b'} &{'a'} #{'a'}

# As both sets contain AC/DC and Back in Black we represent these common elements with the intersection of two circles.
# You can find the intersect of two sets as follow using &

intersection = album_set1 & album_set2
intersection #{'AC/DC', 'Back in Black'}

# Use intersection method to find the intersection of album_list1 and album_list2

album_set1.intersection(album_set2)   #{'AC/DC', 'Back in Black'}
```

```python
# difference 

# You can find all the elements that are only contained in album_set1 (but not in album_set2) using the difference 
#   method

album_set1.difference(album_set2)  #{'Thriller'}
```

```python
# The elements in album_set2 but not in album_set1 is given by:

album_set2.difference(album_set1)  #{'The Dark Side of the Moon'}
```

```python
# union -The union corresponds to all the elements in both sets, which is represented by coloring both circles

album_set1.union(album_set2)
```

```python
# Superset

# is album_set1 superset of album_set2
album_set1.issuperset(album_set2) #False 

# is album_set1 superset os set {"Back in Black", "AC/DC"}
album_set1.issuperset({"Back in Black", "AC/DC"}) #True
```

```python
# Check if subset

# is album_set2 a subset of album_set1
album_set2.issubset(album_set1) #False

# is set {"Back in Black", "AC/DC"} a subset of album_set1
set({"Back in Black", "AC/DC"}).issubset(album_set1) #True 
```
