```python
# Dictionary Comprehensions 

planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune'] # List 

planetToInitial = {planet: planet[0] for planet in planets}
planetToInitial
```

```python
# in operator (more info in operators notebook)

# We can verify if an element is in the dictionary using the "in" command as follows: The command checks the keys. 
#   If they are in the dictionary, they return a true.
# If we try the same command with a key that is not in the dictionary, we get a false.
'Saturn' in planetToInitial #True
'Betelgeuse' in planetToInitial #False
```

```python
# Method - get
# get - looks up values in a dictionary. returns None if the key is not found (or a default value of your choice) 

elements = {"hydrogen": 1, "helium": 2, "carbon": 6}
#look up usisng square brackets 
#print(elements['dilithium']) #KeyError: 'dilithium'

#look up using get method 
print(elements.get("helium")) #2
print(elements.get("dilithium")) #None 
elements.get('kryptonite', 'There\'s no such element!') #"There's no such element!"
```

```python
# for loop - A for loop over a dictionary will loop over its keys
# dict1 = {'key1': 'val1', 'key2': 'val2', 'key3': 'val3', 'key4': 'Pluto'}

for k in dict1:
    print("{} = {}".format(k, dict1[k]))
```

```python
# We can access a collection of all the keys or all the values with dict.keys() and dict.values(), respectively.

planetToInitial.values() # dict_values(['M', 'V', 'E', 'M', 'J', 'S', 'U', 'N'])
planetToInitial.keys() # dict_keys(['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune'])
```

```python
# Get all the initials, sort them alphabetically, and put them in a space-separated string.

'-'.join(sorted(planetToInitial.values()))
```

```python
# dict.items() method lets us iterate over the keys and values of a dictionary simultaneously. (In Python jargon, an 
#   item refers to a key, value pair)

# print(planetToInitial.items()) # dict_items([('Mercury', 'M'), ('Venus', 'V'), ('Earth', 'E'), ('Mars', 'M'), ('Jupiter', 'J'), ('Saturn', 'S'), ('Uranus', 'U'), ('Neptune', 'N')])

for planet, initial in planetToInitial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```
