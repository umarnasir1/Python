```python
# We can include containers in other containers to create compound data structures. For example, this dictionary maps 
#   keys to values that are also dictionaries!

elements = {"hydrogen": {"number": 1,
                         "weight": 1.00794,
                         "symbol": "H"},
              "helium": {"number": 2,
                         "weight": 4.002602,
                         "symbol": "He"}}

# We can access elements in this nested dictionary like this.
elements['helium'] #{'number': 2, 'weight': 4.002602, 'symbol': 'He'}
# get hydrogens weight 
elements['hydrogen']['weight'] #1.00794

# Adding elements to nested data structure 
elements['hydrogen']['is_noble_gas'] = False
elements['helium']['is_noble_gas'] = True

# You can also add a new key to the element dictionary.
oxygen = {"number":8,"weight":15.999,"symbol":"O"}  # create a new oxygen dictionary 
elements["oxygen"] = oxygen  # assign 'oxygen' as a key to the elements dictionary
print('elements = ', elements)
```

```python
# Tuples-Nesting 
# A tuple can contain other tuples as well as other complex data types. This is called nesting.

NT = (1, 2, ("pop", "rock"), (3,4), ("disco",(1,2)))
#     0  1         2           3            4
# It is helpful to visualize this as a tree.

#  We can access these elements using the standard indexing methods. If we consider indexes with other tuples, we see 
#   the tuple at index 2 contains a tuple with two elements. We can access those two indexes.
print("Element 0 of Tuple: ", NT[0]) #Element 0 of Tuple:  1
print("Element 1 of Tuple: ", NT[1]) #Element 1 of Tuple:  2
print("Element 2 of Tuple: ", NT[2]) #Element 2 of Tuple:  ('pop', 'rock')
print("Element 3 of Tuple: ", NT[3]) #Element 3 of Tuple:  (3, 4)
print("Element 4 of Tuple: ", NT[4]) #Element 4 of Tuple:  ('disco', (1, 2))

#  we could use the second index to access other tuples.
print("Element 2, 0 of Tuple: ",   NT[2][0]) #Element 2, 0 of Tuple:  pop
print("Element 2, 1 of Tuple: ",   NT[2][1]) #Element 2, 1 of Tuple:  rock
print("Element 3, 0 of Tuple: ",   NT[3][0]) #Element 3, 0 of Tuple:  3
print("Element 3, 1 of Tuple: ",   NT[3][1]) #Element 3, 1 of Tuple:  4
print("Element 4, 0 of Tuple: ",   NT[4][0]) #Element 4, 0 of Tuple:  disco
print("Element 4, 1 of Tuple: ",   NT[4][1]) #Element 4, 1 of Tuple:  (1, 2)

NT[4] #('disco', (1, 2))
NT[4][0] #'disco'
NT[4][1] #(1, 2)

# we can access elements nested deeper in the tree with a third index
# We can even access deeper levels of the tree by adding another square bracket. We can access different characters in 
#   the string or various elements in the second tuple contained in the first.
NT[2][1][0] #'r'
NT[2][1][1] #'o'

NT[4][1][0] #1
NT[4][1][1] #2
```
