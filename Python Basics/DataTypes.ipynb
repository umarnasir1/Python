## Comparison Operators (==, != , <, >, <=, >=)

Comparison Operators that are commonly used to obtain bool value.   
When comparing two values you can use these operators:
* equal: ==
* not equal: !=
* greater than: >
* less than: <
* greater than or equal to: >=
* less than or equal to: <=

```python
#logical/boolean
# allows us to work with loops, conditional statements & evaluate logical expressions

#Boolean 
# used to represent Logical values, i.e., True/False 
# The bool data type holds one of the values True or False, which are often encoded as 1 or 0, respectively.

4 < 5 #True
10 > 100 #False
4 == 5 #False
# Use Equality sign to compare the strings
"ACDC" == "Michael Jackson" #False
1=2 #SyntaxError:can't assign to literal

# Inequality Sign
# The inequality test uses an exclamation mark preceding the equal sign, if two operands are not equal then the 
#   condition becomes True.
i = 2
i != 6 #True
i != 2 #False
"ACDC" != "Michael Jackson" #True

# if we cast a Boolean True to an integer or float, we will get a 1. If we cast a Boolean False to an integer or float,
#   we get a 0.
int(True) #1 
int(False) #0

# Similarly 
bool(1) # True
bool(0) # False


result = 4 < 5
print(result) #True
type(result) #bool# if we cast a Boolean True to an integer or float, we will get a 1. If we cast a Boolean False to an integer or float,
#   we get a 0.
int(True) #1 
int(False) #0

# Similarly 
bool(1) # True
bool(0) # False


result = 4 < 5
print(result) #True
type(result) #bool
```
